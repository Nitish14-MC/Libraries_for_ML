# Axes labels, Legend, Grid

```python
days=[1,2,3,4,5,6,7]
max_t=[50,51,52,48,47,49,46]
min_t=[43,42,40,44,33,35,37]
avg_t=[45,48,48,46,40,42,41]
```

```python
plt.plot(days,max_t)
```

```python
plt.plot(days,max_t)
plt.plot(days,min_t)
plt.plot(days,avg_t)
```

```python
plt.xlabel("Days")
plt.ylabel("Temperature")
plt.title("Weather details")
plt.plot(days,max_t)
plt.plot(days,min_t)
plt.plot(days,avg_t)
```

# Using Legend property

```python
plt.xlabel("Days")
plt.ylabel("Temperature")
plt.title("Weather details")
plt.plot(days,max_t,label="Max")
plt.plot(days,min_t,label="Min")
plt.plot(days,avg_t,label="Avg")
plt.legend()
```

### Changing the legend to upper right corner

```python
plt.xlabel("Days")
plt.ylabel("Temperature")
plt.title("Weather details")
plt.plot(days,max_t,label="Max")
plt.plot(days,min_t,label="Min")
plt.plot(days,avg_t,label="Avg")
plt.legend(loc="upper right")
```

### PLacing the legend in the most and appropriate position using "BEST"

```python
plt.xlabel("Days")
plt.ylabel("Temperature")
plt.title("Weather details")
plt.plot(days,max_t,label="Max")
plt.plot(days,min_t,label="Min")
plt.plot(days,avg_t,label="Avg")
plt.legend(loc="best")
```

```python
plt.xlabel("Days")
plt.ylabel("Temperature")
plt.title("Weather details")
plt.plot(days,max_t,label="Max")
plt.plot(days,min_t,label="Min")
plt.plot(days,avg_t,label="Avg")
plt.legend(loc="best",shadow=True)
```

```python
plt.xlabel("Days")
plt.ylabel("Temperature")
plt.title("Weather details")
plt.plot(days,max_t,label="Max")
plt.plot(days,min_t,label="Min")
plt.plot(days,avg_t,label="Avg")
plt.legend(loc="best",shadow=True,fontsize='large')
```

```python
# Legend at different location with shadow enabled and fontsize set to large
plt.plot(days, max_t, label="max")
plt.plot(days, min_t, label="min")
plt.plot(days, avg_t, label="average")

plt.legend(loc='upper right',shadow=True,fontsize='large')
```

# Grid

```python
# Legend at different location with shadow enabled and fontsize set to large
plt.plot(days, max_t, label="max")
plt.plot(days, min_t, label="min")
plt.plot(days, avg_t, label="average")

plt.legend(loc='upper right',shadow=True,fontsize='large')
plt.grid()
```
